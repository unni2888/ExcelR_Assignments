```python
# Question 12
import numpy as np
import statistics as st
a=[34,36,36,38,38,39,39,40,40,41,41,41,41,42,42,45,49,56]
print('mean =',np.mean(a))
print('median =',np.median(a))
print('mode =',st.mode(a))
print('variance =',np.var(a))
print('std =',np.std(a))
```
