# K-Means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, completeness_score
```

```python
scaler = StandardScaler()
scaled_air3 = scaler.fit_transform(air3)
scaled_air3
```

```python
wcss=[]
for i in range(1,21):
  model = KMeans(n_clusters = i, random_state = 42)
  model.fit(scaled_air3)
  wcss.append(model.inertia_)
```

```python
fig = plt.figure(figsize = (20,6))
plt.plot(range(1,21),wcss)
plt.title("Elbow plot for KMeans")
plt.xlabel('K Values')
plt.xticks(range(21))
plt.ylabel('WCSS')
plt.show()
```

```python
final_K_model = KMeans(n_clusters = 5, random_state = 42)
final_K_model.fit(scaled_air3)
K_labels = final_K_model.labels_
```

```python
K_labels
```

# Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
```

```python
air4 = air3.drop(['Labels'], axis = 1)
```

```python
minmax = MinMaxScaler()
scaled_air4 = minmax.fit_transform(air4)
scaled_air4
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score



# Initialize variables
max_clusters = 10
silhouette_scores = np.zeros(max_clusters-1)

# Iterating over different numbers of clusters
for n_clusters in range(2, max_clusters+1):
    # Performing agglomerative clustering
    hier_model = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean', linkage='complete')
    hier_model.fit(scaled_air4)
    
    # Calculating silhouette score
    silhouette_avg = silhouette_score(scaled_air4, hier_model.labels_)
    silhouette_scores[n_clusters-2] = silhouette_avg
```

```python
plt.plot(range(2, max_clusters+1), silhouette_scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

```python
hier_model = AgglomerativeClustering(n_clusters = 3,  affinity = 'euclidean', linkage = 'complete')
hier_model.fit(scaled_air4)
```

```python
hier_labels = hier_model.labels_
hier_labels
```

```python
air4['Hier_labels'] = hier_labels
air4.head()
```

```python
grp2 = air4.groupby(['Hier_labels'])
```

```python
for x,y in grp2:
  print(y)
  print('_________________________________________________________________________')
```

```python
grp2.agg(['mean'])
```

```python
plt.scatter(air4['Hier_labels'], air4.index, c= hier_labels, cmap = 'viridis')
plt.xticks([0,1,2])
plt.show()
```

# DBSCAN

```python
air5 = air4.drop(['Hier_labels'], axis = 1)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db_scaler = StandardScaler()
scaled_air5 = db_scaler.fit_transform(air5)
scaled_air5
```

```python
eps=np.linspace(0.1,1,20)
min_samples=range(2,20)
sil=[]
for i in eps:
    for j in min_samples:
        cluster=DBSCAN(eps=i,min_samples=j)
        label=cluster.fit_predict(scaled_air5)
        if len(set(label))>1:
             sil.append([i,j,silhouette_score(scaled_air5,label)])
```

```python
eps, min_samples, score = sorted(sil, key=lambda x:x[-1])[-1]
print(f"Best silhouette_score: {score}")
print(f"min_samples: {min_samples}")
print(f"eps: {eps}")
```

```python
dataframe = pd.DataFrame(sil, columns=['eps','min','silhouette_score']).sort_values(['silhouette_score'],ascending = False)
dataframe
```

```python
db_model = DBSCAN(eps = 1, min_samples= 8)
db_model.fit(scaled_air5)
```

```python
db_labels = db_model.labels_
db_labels
```

```python
silhouette_score(air5.iloc[:,:10], db_labels)
```
